```python
import pandas as pd
import random as rd
lst = ['robot'] * 10
lst += ['human'] * 10
rd.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
list = []
len = data.shape[0]
columns = data['whoAmI'].unique()
for i in range(len):
  if data['whoAmI'][i] == columns[0]:
    list.append([1,0])
  else:
    list.append([0,1])
df = pd.DataFrame(data= list,columns=columns)
df
```
